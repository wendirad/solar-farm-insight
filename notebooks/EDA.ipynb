## **Dataset Overview**
The dataset provides time-series data to analyze solar energy potential and environmental conditions, aiding in site selection for solar farm installations.

### **Extract dataset**

```python
# Dataset path
RAW_DATASET_PATH = pathlib.Path("../data/data.zip")
EXTRACTION_OUTPUT_PATH = RAW_DATASET_PATH.parent.parent

DATASET_PATH = RAW_DATASET_PATH.parent
```

```python
# Extract the data
extract_data(RAW_DATASET_PATH, EXTRACTION_OUTPUT_PATH)

# Verify intallation
list((EXTRACTION_OUTPUT_PATH / "data").glob("*.csv"))
```

### **Load the dataset**

```python
# Load Benin data
bn_df = pd.read_csv(
    DATASET_PATH / "benin-malanville.csv", parse_dates=["Timestamp"]
)

# Load Sierraleone data
sl_df = pd.read_csv(
    DATASET_PATH / "sierraleone-bumbuna.csv", parse_dates=["Timestamp"]
)

# Load Togo data
tg_df = pd.read_csv(
    DATASET_PATH / "togo-dapaong_qc.csv", parse_dates=["Timestamp"]
)
```
